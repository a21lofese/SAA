## Cargar el dataset

```python
import pandas as pd
df = pd.read_csv("csv/rl1.csv")
df.head()
```

## Preparamiento de los datos

```python
X, y = df[['x']], df['y']
```

## Visualizaci√≥n de los datos

```python
from matplotlib import pyplot as plt

plt.scatter(df['x'], df['y'], label='Datos')
plt.title('Distribuci√≥n de los datos')
plt.show()
```

```python
import seaborn as sns

sns.lmplot(data=df, x='x', y='y', order=1)
```

## Preparaci√≥n del conjunto de datos

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=42)
```

## Creaci√≥n del modelo

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

print(f"Coeficiente: {model.coef_}")
print(f"Intercepto: {model.intercept_}")
```

## Predicci√≥n

```python
y_pred = model.predict(X_test)

from sklearn import metrics

print(f"MSE: {metrics.mean_squared_error(y_test, y_pred)}")
print(f"R2: {metrics.r2_score(y_test, y_pred)}")
print(f"var: {metrics.explained_variance_score(y_test, y_pred)}")
```

## Visualizaci√≥n de la predicci√≥n

```python
plt.scatter(X_test, y_test, label='Datos')
plt.plot(X_test, y_pred, color='red', label='Predicci√≥n')
plt.title('Predicci√≥n')
plt.legend()
plt.show()
```
